```python
import pandas as pd
import numpy as np
import chart_studio.plotly as py
import cufflinks as cf
import seaborn as sns
import plotly.express as px
%matplotlib inline

# Make Plotly work in your Jupyter Notebook
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
# Use Plotly locally
cf.go_offline()

#Dash
import dash
import dash_core_components as dcc
from dash import html
from dash.dependencies import Input, Output
import re
```

```python
file_path = "/Users/jeremyrodriguez/Desktop/PMC Data/PMC Project/TrackingMain.csv"

df= pd.read_csv(file_path)


#Pull Location and Direction. Will be useful for having the total volume over time
df['Location'] = df['Location In'].str.replace(r' in| out', '', case=False, regex=True)
df['Date_Time'] = pd.to_datetime(df['Date'] + ' ' + df['Hour'].astype(str) + ':00:00')
df=df.drop('Unnamed: 0',axis = 1)
```

```python
df
```

```python

import matplotlib.pyplot as plt

in_entries = df[df['Location In'].str.contains('In', case = False)]
out_entries = df[df['Location In'].str.contains('Out', case = False)]

unique_locations = df['Location In'].str.extract(r'(.*?)( In| Out|$)')[0].unique()
water_stop_data = df[df['Location'].str.contains('Whitinsville')]

# Calculate rider volume over time
time_intervals = water_stop_data['Date_Time']
rider_volume = []
volume = 0
for time in time_intervals:
    in_count = len(in_entries[in_entries['Date_Time'] == time])
    out_count = len(out_entries[out_entries['Date_Time'] == time])
    volume += in_count - out_count
    rider_volume.append(volume)

# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(time_intervals, rider_volume)
plt.xlabel('Time')
plt.ylabel('Rider Volume')
plt.title('Rider Volume Over Time at Whitinsville Water Stop')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='location-dropdown',
        options=[{'label': i, 'value': i} for i in df['Location In'].unique()],
        value='Location1'  # Default value
    ),
    dcc.Graph(id='riders-plot')
])

# Define callback to update graph
@app.callback(
    Output('riders-plot', 'figure'),
    [Input('location-dropdown', 'value')]
)
def update_graph(selected_location):
    filtered_df = df[df['Location In'] == selected_location]
    fig = px.line(filtered_df, x='time', y='riders',labels={'riders': 'Number of Riders', 'Hour': 'Hour'},
                 title=f'Riders per Hour at {selected_location}')
    return fig


# Run the app
if __name__ == '__main__':
    app.run_server(mode='external')  # Use 'external' if you want to open the app in a new browser tab
```
